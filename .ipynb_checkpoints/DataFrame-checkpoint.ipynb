```python
# fillna : used to fill a value in every none value in dataframe

A = [[1,2,3,4,5],[6,7,8,9,0],[None,None,None,None,None]]
a = pd.DataFrame(A, columns=['s1','s2','s3','s4','s5'])
# a.fillna(0)         default filling
print(a['s3'].fillna(5))     # filling at a specific position
# a['s3'] = a['s3'].fillna(5)    for permanent change
print(a)
print(a['s4'].fillna(a['s4'].mean()))    # replacing with mean value
print(a['s5'].ffill())
# print(a['s5'])
a['s5'].bfill()
```

# apply function

```python
# user defined

def ab(n):
    if n>10:
        return n
```

```python
df = pd.read_csv('myrgroup.csv')
df['windspeed'].apply(ab)
```

```python
# pre defined function
df = pd.read_csv('myrgroup.csv')
df['windspeed'] = df['windspeed'].fillna(0)
df['windspeed'] = df['windspeed'].astype(int)
df.apply(min)
```

```python
# lambda function
df = pd.read_csv('myrgroup.csv')
result = df['temp'].apply(lambda x : x+10)
result
```

## group by

```python
df = pd.read_csv('myrgroup.csv')
print(df[df['city'].isin(['bhopal','ujjain'])])
c = df.groupby('city')

for x,y in c:
    print(x)
    print(y)
c.get_group('bhopal')
```

```python
df = pd.read_csv('myrgroup.csv')
print(df[df['city'].isin(['bhopal','ujjain'])])
print(df[~df['city'].isin(['bhopal','ujjain'])])
```

```python
df1 = pd.DataFrame({ 'A':[1,2,3,4],'B':[5,6,7,8]})
df2 = pd.DataFrame({'C':[10,20,30,40],'D':[50,60,70,80]})
# join
df1.join(df2)
```

```python
df1 = pd.DataFrame({ 'A':[1,2,3,4],'B':[5,6,7,8]})
df2 = pd.DataFrame({'C':[10,20],'D':[50,60]})
# in the case of uneven no.of rows in both table, missing values are replaced by NaN
df1.join(df2)
```

```python
df1 = pd.DataFrame({ 'A':[1,2,3,4],'B':[5,6,7,8]}, index=['a','b','c','d'])
df2 = pd.DataFrame({'C':[10,20,30,40],'D':[50,60,70,80]})
# case: if index for one of the dataframe is given then join is performed as left join meaning data of only left table is returned and other values are returned as NaN
df1.join(df2)
```

```python
df1 = pd.DataFrame({ 'A':[1,2,3,4],'B':[5,6,7,8]}, index=['a','b','c','d'])
df2 = pd.DataFrame({'C':[10,20,30,40],'D':[50,60,70,80]}, index=['e','f','g','h'])
df1.join(df2)
```

```python
df1 = pd.DataFrame({ 'A':[1,2,3,4],'B':[5,6,7,8]}, index=['a','b','c','d'])
df2 = pd.DataFrame({'C':[10,20,30,40],'D':[50,60,70,80]}, index=['a','b','x','d'])
print('inner join', df1.join(df2, how='inner'))
print('left join', df1.join(df2, how='left'))
print('right join', df1.join(df2, how='right'))
```

```python
df5 = pd.DataFrame({ 'A':[1,2,3,4],'B':[5,6,7,8]}, index= ['a','b','c','d'])
df6 = pd.DataFrame({'A':[10,20],'D':[50,60]},index=['a','b'])
df5.join(df6, how='left', lsuffix='left', rsuffix='right')
```

```python
# concat 
d1 = pd.DataFrame([100,200,300,400])
d2 = pd.DataFrame([500,600,700,800])
d3 = pd.DataFrame([900,1000,1100,1200])
d = pd.concat([d1,d2,d3])
d
```

```python
df1 = pd.DataFrame({'A':[10,20],'B':[30,40]})
df2 = pd.DataFrame({'A':[50,60], 'B':[70,80]})
df = pd.concat([df1,df2],ignore_index=True)
df
```
