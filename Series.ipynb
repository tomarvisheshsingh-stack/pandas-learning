```python
# methods of series

# head() : gives starting values of data (by default 5)
score = [76,65,88,96,45,78,55,34,12,2,42,73,98]
s = pd.Series(score)
print(s.head(5))
```

```python
# tail() : gives last values of data (by default 5)
score = [76,65,88,96,45,78,55,34,12,2,42,73,98]
s = pd.Series(score)
print(s.tail(2))
```

```python
# sample() : gives random value from data (by default 1)
score = [76,65,88,96,45,78,55,34,12,2,42,73,98]
s = pd.Series(score)
print(s.sample(5))
```

```python
# null variation
score = [76,np.nan,88,96,None,78,65,34,None,2,42,73,98]
s = pd.Series(score)
# print(s)
# .isnull() : gives true at positions where value is null
print(s.isnull())
# .hasnans : gives true if there's none value in series
print(s.hasnans)
# .notnull : gives true at positions where value is not null
print(s.notnull())
```

```python
score = [76,np.nan,88,96,None,78,65,34,None,2,42,73,98]
s = pd.Series(score)
print(s.isnull().sum())
```

```python
# membership operator : seraches index

s = pd.Series([10,215,54,21,65423,11,2,223,44])
print(s)
print(10 in s)
```

```python
s = pd.Series([10,215,54,21,65423,11,2,223,44])
print(s)
print(145 in s.values)
```

```python
# looping 
s = pd.Series([10,215,54,21,65423,11,2,223,44])
print(s)
for i in s:
    print(i)
```

```python
s = pd.Series([10,215,54,21,65423,11,2,223,44])
print(s)
for i in range(len(s)):
    s[i] = s[i] + 10
print(s)
```

```python
# arthematic operator

# Broadcasting addition
s = pd.Series([10,215,54,21,65423,11,2,223,44])
print(s)
s = s + 10
print(s)
s = s - 10
print(s)
```

```python
s = pd.Series([10,215,54,21,65423,11,2,223,44])
for i in s:
    if i>15:
        print(i)
print()
print(s[s>15])
print()
print(s)
```
