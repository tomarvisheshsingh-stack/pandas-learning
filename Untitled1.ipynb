```python
# Q.4 Write a Pandas program to filter words from a given series that contain atleast two vowels.
s = pd.Series(['apple', 'sky', 'banana', 'try', 'education'])

a = 'aieou'
for i in s:
    count = 0
    for j in i:
        if j in a:
            count += 1
    if count >= 1:
        print(i)
```
