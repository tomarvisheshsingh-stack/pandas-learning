```python
a = [1,2,3,4,5]
print(a, type(a))
k=pd.Series(a)
print(k, type(k))
```

```python
a = { 'a':10, 'b':12, 'c':13}
print(a, type(a))
k=pd.Series(a)
print(k, type(k))
```

```python
# attributes of series
a = [1,2,3,4,5]
print(a, type(a))
k=pd.Series(a)
print(k, type(k))
print(k.size)
print(k.dtype)
print(k.name)
print(k.index)
print(k.values)
```

```python
a = [1,2,3,4,5, True]
print(a, type(a))
k=pd.Series(a, name = 'lit')
print(k, type(k))
print(k.dtype)
print(k.name)
print(k.is_unique)
```

```python
a = {'name':['a','b'], 'age':[10,20]}
k=pd.Series(a)
print(k)
print(k['name'][1])
```

```python
name = ['phy', 'chem', 'math', 'eng']
score = [76,65,88,96]
s = pd.Series(score, index = name)
print(s)
```

```python
# methods of series

# head() : gives starting values of data (by default 5)
score = [76,65,88,96,45,78,55,34,12,2,42,73,98]
s = pd.Series(score)
print(s.head(5))
```

```python
# tail() : gives last values of data (by default 5)
score = [76,65,88,96,45,78,55,34,12,2,42,73,98]
s = pd.Series(score)
print(s.tail(2))
```

```python
# sample() : gives random value from data (by default 1)
score = [76,65,88,96,45,78,55,34,12,2,42,73,98]
s = pd.Series(score)
print(s.sample(5))
```

```python
score = [76,65,88,96,65,78,65,34,2,2,42,73,98]
s = pd.Series(score)
print(s.value_counts())
```

```python
score = [76,65,88,96,45,78,55,34,12,2,42,73,98]
s = pd.Series(score)
print(s.sort_values())
```

```python
s = ['a','p','p','l','e']
a = pd.Series(s)
print(a.sort_values())
print(a.sort_values(ascending=False))
```
