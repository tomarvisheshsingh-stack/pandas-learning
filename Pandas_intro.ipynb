```python
a = [1,2,3,4,5]
print(a, type(a))
k=pd.Series(a)
print(k, type(k))
```

```python
a = { 'a':10, 'b':12, 'c':13}
print(a, type(a))
k=pd.Series(a)
print(k, type(k))
```

```python
# attributes of series
a = [1,2,3,4,5]
print(a, type(a))
k=pd.Series(a)
print(k, type(k))
print(k.size)
print(k.dtype)
print(k.name)
print(k.index)
print(k.values)
```

```python
a = [1,2,3,4,5, True]
print(a, type(a))
k=pd.Series(a, name = 'lit')
print(k, type(k))
print(k.dtype)
print(k.name)
print(k.is_unique)
```

```python
a = {'name':['a','b'], 'age':[10,20]}
k=pd.Series(a)
print(k)
print(k['name'][1])
```
